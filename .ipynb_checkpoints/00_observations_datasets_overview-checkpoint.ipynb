# Preprocessed observational datasets for tropical cyclone evaluation

This notebook is **only about observations** (no WRF output here).

We use three observational datasets:

### 1) IBTrACS / best track (here: `matthew_obs.nc`)
Provides the *best-estimate* storm evolution:
- center position (**lat**, **lon**)
- minimum sea-level pressure (**usa_pres**)
- maximum sustained wind (**usa_wind**)
- https://www.ncei.noaa.gov/products/international-best-track-archive

### 2) IMERG precipitation (here: `IMERG_3h_on_WRF.nc`)
Satellite-based precipitation estimates that are ground callibrated.
In our course file, IMERG has already been **interpolated to the WRF grid** and provides:
- **precipitation** (mm per 3h)
- coordinates: **XLAT**, **XLONG**, **time**
- https://gpm.nasa.gov/resources/documents/imerg-v07-release-notes

### 3) Merged IR brightness temperature (here: `Tb_on_wrf.nc`)
Geostationary IR observations, already mapped to the WRF grid.
In our course file, the variable is:
- **Tb** (brightness temperature, K)
- coordinates: **XLAT**, **XLONG**, **time**
- https://disc.gsfc.nasa.gov/datasets/GPM_MERGIR_1/summary
